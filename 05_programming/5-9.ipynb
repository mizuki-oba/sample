```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[0])  # a（先頭）
print(s[-1])  # z（末尾）
```

```python
s = "abcdefghijklmnopqrstuvwxyz"
print(s[0:7])    # abcdefg
print(s[-5:])    # vwxyz
print(s[::2])    # acegikmoqsuwy（2個おき）
```

```python
s = "TechAcademy"
for char in s:
    print(char.upper())  # 各文字を大文字に変換して表示
```

```python
s = "TechAcademy"
print(len(s))
if len(s) > 10:
    print("長い文字列です。")
```

```python
s = "にんてんどううどんてんに"
if s == s[::-1]:
    print(f"「{s}」は回文です！")
else:
    print(f"「{s}」は回文ではありません。")
```

```python
numbers = [1, 2, 3, 4, 5]
fruits = ["apple", "banana", "cherry"]
mixed = [1, "apple", 3.14, True]
empty_list = []

print(numbers, fruits, mixed, empty_list)
```

```python
string_to_list = list("hello")
tuple_to_list = list((1, 2, 3))
range_to_list = list(range(5))

print(string_to_list, tuple_to_list, range_to_list)
```

```python
none_list = [None] * 5
zeros = [0] * 10

print(none_list, zeros)
```

```python
original = [1, 2, 3]
copy1 = original[:]
copy2 = list(original)
copy3 = original.copy()

print(copy1, copy2, copy3)
```

```python
numbers = [0, 1, 2, 3, 4, 5]
print(numbers[1:3], numbers[:3], numbers[2:], numbers[::-1])
```
